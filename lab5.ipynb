```python
import numpy as np

string1 = "hi my name Python"
string2 = "python is very fast"

print(np.char.add(string1, string2))
print()

print(np.char.multiply([string1, string2], [2, 5]))
print()

print(np.char.center(string1, 50, "*"))
print()

print(np.char.capitalize([string1, string2]))
print()

print(np.char.title([string1, string2]))
print()

print(np.char.upper([string1, string2]))
print()

print(np.char.lower([string1, string2]))
print()

print(np.char.upper([string1, string2]))
print()

print(np.char.replace([string1, string2], ['hi', 'is'], ['hello', 'was']))
print()

print(np.char.join('---', [string1, string2]))
print()
```

```python
arr1 = np.random.randint(100, size=(4,4))
arr2 = np.random.randint(100, size=(4,4))

arr3 = np.multiply(arr1, arr2)
print('arr2*arr1:\n' ,arr3, '\n')

arr4 = np.divide(arr3, arr1)
print('(arr2*arr1)/arr1:\n', arr4, '\n')

arr5 = np.add(arr1, arr4)
print('arr1+((arr2*arr1)/arr1):\n', arr5, '\n')

arr6 = np.subtract(arr5, arr1)
print('arr1+((arr2*arr1)/arr1)-arr1:\n', arr6, '\n')
```

```python
array = np.random.randint(20, size=(6,6))

print('array:\n', array, '\n')

print('Sorted array:')
print(np.sort(array))
print()

print('Sorted array by rows:')
print(np.sort(array, axis=0))
print()

print('Sorted array by columns:')
print(np.sort(array, axis=1))
print()

print("Row's and Column's index where 8 is found:")
print(np.array(np.where(array == 8)))
print()

print("Min and Max:", end = " ")
print(np.min(array), end = " ")
print(np.max(array))
print()

print("Min and Max by columns:", end = " ")
print(np.min(array, axis=0), end = " ")
print(np.max(array, axis=0))
print("Index of min and max by columns:", end = " ")
print(np.argmin(array, axis=0), end = " ")
print(np.argmax(array, axis=0))
print()

print("All elements are zero:", end = " ")
print(np.all(array == 0))
print()

print("Any elements are zero:", end = " ")
print(np.any(array == 0))
print()
```

```python
large_array = np.random.randint(100, size=(4032000))
print("Large array:")
print(large_array.reshape(1920, 2100))
print()

sub_arrs = int(1920*2100/120/100)
print(f"Splitting the array into {sub_arrs} sub-arrays:")
splitted_array = np.split(large_array, sub_arrs)
for a in splitted_array:
    print(np.array(a).reshape(120, 100))
    print()
```

```python
int_array = np.random.randint(20, size=(4,2))
print("Integer array:")
print(int_array)
print()

print("class: ", type(int_array))
print("datatype: ", int_array.dtype)
print("shape: ", int_array.shape)
print("size: ", int_array.size)
print("no. of bytes: ", int_array.nbytes)
print("dimension: ", int_array.ndim)
print()
```

```python
stat_array = np.random.randint(30, size=(4,4))
print("Array:")
print(stat_array)
print()

print("Row-wise Mean:", end = " ")
print(np.mean(stat_array, axis=1))
print("Column-wise Mean:", end = " ")
print(np.mean(stat_array, axis=0))
print()

print("Row-wise Median:", end = " ")
print(np.median(stat_array, axis=1))
print("Column-wise Median:", end = " ")
print(np.median(stat_array, axis=0))
print()

print("Row-wise Variance:", end = " ")
print(np.var(stat_array, axis=1))
print("Column-wise Variance:", end = " ")
print(np.var(stat_array, axis=0))
print()
```
