### Question 1

```python
def save_csv(df: pd.DataFrame):
    temp_df = df.reset_index()
    temp_df.set_index('Name', inplace=True)
    temp_df.to_csv("/content/drive/MyDrive/DS-Lab-CS21B2018/SBIAccountHolder.csv")

sbi_acc = {
    "Account_No": [],
    "Name": [],
    "Account_Type": [],
    "Adhaar_No": [],
    "Balance": []
}

acc_df = pd.DataFrame(sbi_acc)
save_csv(acc_df)
print()
print("Data saved as 'SBIAccountHolder.csv': ")
print(acc_df)
acc_df.set_index("Account_No", inplace=True)
```

```python
print("1. Append record")
print("2. Delete Record")
print("3. Credit from the account")
print("4. Debit from the account")
print("5. Print account details")
print("6. Exit")
choice = 0

while choice != 6:
    choice = int(input("\nEnter Choice: "))

    if choice == 1:
        print("New Entry:")
        new_acc = int(input("\tAccount Number: "))

        new_entry = []
        new_entry.append(input("\tName: "))
        new_entry.append(input("\tAccount Type: "))
        new_entry.append(input("\tAdhaar No: "))
        new_entry.append(int(input("\tBalance: ")))

        acc_df.loc[new_acc] = new_entry
        save_csv(acc_df)
    
    elif choice == 2:
        acc_num = int(input("Enter the account number to delete: "))

        try:
            acc_df.drop(acc_num, inplace=True)
        except KeyError:
            print(f"Account no '{acc_num}' does not exist")
        
        save_csv(acc_df)
    
    elif choice == 3:
        acc_num = int(input("Enter the account number: "))

        amnt = int(input("\t Credit amount: "))
        if amnt < 0:
            print(f"Invalid Amount '{amnt}'")
            continue
        
        try:
            acc_df.loc[acc_num, "Balance"] += amnt
        except KeyError:
            print(f"Account no '{acc_num}' does not exist")
        
        save_csv(acc_df)
    
    elif choice == 4:
        acc_num = int(input("Enter the account number: "))

        amnt = int(input("\t Debit amount: "))
        if amnt < 0:
            print(f"Invalid Amount '{amnt}'")
            continue
        
        try:
            if acc_df.loc[acc_num, "Account_Type"] == 'SB' and amnt > acc_df.loc[acc_num, "Balance"]:
                print("Insufficient Balance")
            else:
                acc_df.loc[acc_num, "Balance"] -= amnt
        except KeyError:
            print(f"Account no '{acc_num}' does not exist")
        
        save_csv(acc_df)
    
    elif choice == 5:
        acc_num = int(input("Enter the account number: "))
        try:
            print(acc_df.loc[acc_num:acc_num])
        except KeyError:
            print(f"Account no '{acc_num}' does not exist")

    elif choice == 6:
        print("Exiting...")
    
    else:
        print("Invalid Choice")
```

### Question 2

```python
num = int(input("Enter the number of accounts: "))
adhaar = {
    "Adhaar_No": [],
    "Name": [],
    "Contact_No": [],
    "DOB": [],
    "Address": []
}
for i in range(num):
    print(f"Entry{i+1}:")
    adhaar["Name"].append(input("\tName: "))
    adhaar["Adhaar_No"].append(int(input("\tAdhaar No: ")))
    adhaar["Contact_No"].append(int(input("\tContact No: ")))
    adhaar["DOB"].append(input("\tDOB: "))
    adhaar["Address"].append(input("\tAddress: "))

adhaar_df = pd.DataFrame(adhaar)
adhaar_df.set_index('Adhaar_No', inplace=True)
adhaar_df.to_csv("/content/drive/MyDrive/DS-Lab-CS21B2018/Adhaar_DB.csv")
print()
print("Data saved as 'Adhaar_DB.csv': ")
print(adhaar_df)
```

### Merging two Tables

```python
df1 = pd.read_csv("/content/drive/MyDrive/DS-Lab-CS21B2018/SBIAccountHolder.csv", index_col=None)

print(df1)
```

```python
df2 = pd.read_csv("/content/drive/MyDrive/DS-Lab-CS21B2018/Adhaar_DB.csv", index_col=None)

print(df2)
```

```python
print("Merging both the csv as 'MergeBankAdhaar.csv':")
comb_df = pd.merge(df1, df2, how='outer', on=['Adhaar_No', 'Name'])
print(comb_df)
comb_df.to_csv("/content/drive/MyDrive/DS-Lab-CS21B2018/MergeBankAdhaar.csv")
```
