#### Question 1
Write a Python program to remove an empty tuple(s) from a list of tuples.

```python
n = int(input("Enter the number of tuples in list: "))
lst = []
for i in range(n):
    lst.append(tuple(input(f"Tuple{i+1}: ").split()))
print("List before: ", lst)

try:
    while True:
        lst.remove(tuple())
except ValueError:
    pass

print("List after: ", lst)
```

#### Question 2
Write a Python program to count the elements in a list until an element is a
tuple

```python
n = int(input("Enter the number of elements: "))
lst = []
print("Enter tuple as (), list as [], string as '', int normally")
for i in range(n):
    val = input(f"Element {i+1}: ")
    try:
        val = eval(val)
    except NameError:
        pass
    lst.append(val)
```

```python
print(lst)

count  = 0
for a in lst:
    if type(a).__name__ == "tuple":
        break
    count += 1
print(f"Number of elements until a tuple: {count}")
```

#### Question 3
Write a Python program to convert a tuple of string values to a tuple of integer
values.

```python
tpl = tuple(map(str, input("Enter the elements of tuple: ").split()))
print("Tuple before:", tpl)

tpl = tuple(int(a) for a in tpl)
print("Tuple after:", tpl)
```

#### Question 4
Write a Python program to check if a specified element presents in a tuple of
tuples

```python
tpl = (('Red', 'White', 'Blue'),
       ('Green', 'Pink', 'Purple'),
       ('Orange', 'Yellow', 'Lime'))
print(tpl)
elem = input("Enter the element to search: ")
for t in tpl:
    if elem in t:
        print(True)
        break
else:
    print(False)
```

#### Question 5
Write a Python program to convert a given list of tuples to a list of lists and to a
numpy array.

```python
n = int(input("Enter the number of tuples in the list: "))
lst_of_tpl = []
for i in range(n):
    lst_of_tpl.append(tuple(map(int, input(f"Tuple{i+1}: ").split())))

print("List of tuples:", lst_of_tpl)
lst_of_lst = list(list(t) for t in lst_of_tpl)
print("List of lists:", lst_of_lst)
print("Numpy arrays:")
print(np.array(lst_of_lst))
```

#### Question 6
Write a Python program to remove duplicates from Dictionary.

```python
n = int(input("Enter the number of keys in dictionary: "))
print("Enter Key Value:")
dct = dict(input().split() for i in range(n))
print("Dictionary before:", dct)

dct1 = {}
for key, value in dct.items():
    if value not in dct1.values():
        dct1[key] = value
print("Dictionary after:", dct1)
```

#### Question 7
Write python code to create histogram based on given data:

rcb: 78,63,49,41,68,101,56,79,68,96

mi: 45,85,98,102,42,50,43,48,63,39

Plot the data in different format.

```python
rcb = np.array([78,63,49,41,68,101,56,79,68,96])
bins = [i for i in range(0,np.max(rcb)+11, 10)]
plt.hist(rcb, bins=bins, color="red")
plt.title("RCB")
plt.show()
```

```python
mi = np.array([45,85,98,102,42,50,43,48,63,39])
bins = [i for i in range(0,np.max(mi)+11, 10)]
plt.hist(mi, bins=bins, color="blue")
plt.title("MI")
plt.show()
```
