## Data Collection

```python
file_path = "/content/drive/MyDrive/DS-Lab-CS21B2018/bill_authentication.csv"
bill_df = pd.read_csv(file_path)
bill_df
```

```python
features = bill_df.drop('Class', axis=1)
label = bill_df['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.4, random_state=1)
```

# Naive Bayes

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

start = t.time()
clf=RandomForestClassifier(n_estimators=100)
clf.fit(X_train,y_train)
yr_pred=clf.predict(X_test)
end = t.time()

tr12 = end - start
print(yr_pred)
print(f"Time : {round(end-start,3)}sec")
```

### Finding important features and generating model on the same

```python
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train,y_train)

sns.barplot(x=clf.feature_importances_, y=['Variance', 'Skewness', 'Curtosis', 'Entropy'])
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Visualizing Important Features")
plt.show()
```

```python
imp_feat = ['Variance', 'Skewness', 'Curtosis']
Xnew_train, Xnew_test = X_train[imp_feat], X_test[imp_feat]
```

```python
from sklearn.ensemble import RandomForestClassifier

start = t.time()
clf=RandomForestClassifier(n_estimators=100)
clf.fit(Xnew_train,y_train)
yr2_pred=clf.predict(Xnew_test)
end = t.time()

tr22 = end - start
print(yr2_pred)
print(f"Time : {round(end-start,3)}sec")
```

```python
from sklearn import metrics
a41 = metrics.f1_score(y_test, yr2_pred)
a42 = metrics.recall_score(y_test, yr2_pred)
a43 = metrics.precision_score(y_test, yr2_pred)
a44 = metrics.accuracy_score(y_test, yr2_pred)
print("Random Forest Classifier with selected features model accuracy(in %):",(metrics.accuracy_score(y_test, yr2_pred))*100)
```
