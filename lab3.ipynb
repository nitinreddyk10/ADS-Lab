```python
import itertools
class SumZero:
    def __init__(self, arr):
        self.ls = arr
    def zero(self):
        self.zero_ls = []
        ls_comb = list(itertools.combinations(self.ls, 3))
        for comb in ls_comb:
            if sum(comb) == 0:
                self.zero_ls.append([comb[x] for x in range(3)])
        return self.zero_ls

nums = SumZero(list(map(int, input("Enter set of numbers: ").split())))
print(nums.zero())
```

```python
import math
class Circle:
    def __init__(self, rad):
        self.radius = rad
    def area(self):
        ar = math.pi * self.radius**2
        return ar
    def permtr(self):
        pr = 2 * math.pi * self.radius
        return pr

c1 = Circle(int(input("Enter the radius of circle: ")))
print(f"Area: {c1.area()}")
print(f"Perimeter: {c1.permtr()}")
```

```python
class PairSum:
    def __init__(self, arr):
        self.ls = arr
    def sum_target(self, tar):
        for i in range(len(self.ls)-1):
            if self.ls[i] + self.ls[i+1] == tar:
                return (i, i+1)

nums = PairSum(list(map(int, input("Enter an array of numbers: ").split())))
print(nums.sum_target(int(input("Enter target: "))))
```
