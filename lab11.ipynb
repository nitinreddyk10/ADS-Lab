 Also identify the outliers using box plot.

```python
def outlier(ser: pd.Series):
    q1 = ser.quantile(0.25)
    q3 = ser.quantile(0.75)
    iqr = q3 - q1
    low_bound = q1 - 1.5 * iqr
    high_bound = q3 + 1.5 * iqr
    ser.sort_values(inplace=True)
    
    return ser[(ser < low_bound) | (ser > high_bound)]
```

```python
print("Outliers:\n")
for col in req_diab_df:
    outl_ser = outlier(req_diab_df[col].copy())
    outl_ser.drop_duplicates(inplace=True)
    print(col, ":", outl_ser.to_csv(header=False, index=False).replace('\n', ', '))
```

### 3. Find the mean, median, mode of histogram for each column in cardata.csv. 
We can use the following formula to find the best estimate of the mean of any histogram. Also identify the outliers(if  any). Take the equal width bin


```python
f_path = "/content/drive/MyDrive/DS-Lab-CS21B2018/car data.csv"
car_df = pd.read_csv(f_path)
display(car_df)
```

```python
def get_hist(arr, bin_width):
    max_bin = int(arr.max()) + bin_width
    hist, bins = np.histogram(arr, bins=range(0,max_bin,bin_width))

    return (hist, bins)
```

```python
total = len(car_df)

col = ['Selling_Price', 'Present_Price', 'Kms_Driven']
bin_widths = [2, 2, 5000]
hists, bins = [], []
for i in range(3):
    tmp_hist, tmp_bin = get_hist(car_df[col[i]], bin_widths[i])
    hists.append(tmp_hist)
    bins.append(tmp_bin)
```

Estimate of Mean: Σmi*ni / N

 where:
- mi: The midpoint of the ith bin
- ni: The frequency of the ith bin
- N: The total sample size


```python
def hist_mean(hists, bins):
    total_n = sum(hists)
    mp_bin = [(bins[i] + bins[i+1]) // 2 for i in range(len(bins)-1)]
    mean_sum = sum([hists[i] * mp_bin[i] for i in range(len(hists))])
    mean_val = mean_sum/total_n

    return round(mean_val, 4)
```

```python
for i in range(len(col)):
    print("Mean", col[i], ":", hist_mean(hists[i], bins[i]))
```

Estimate of Median: L + ( (n/2 – F) / f ) * w

where:
- L: The lower limit of the median group
- n: The total number of observations
- F: The cumulative frequency up to the median group
- f: The frequency of the median group
- w: The width of the median group


```python
def hist_median(hists, bins):
    total_n = sum(hists)
    cum_freq = 0
    for i in range(len(hists)):
        # print(cum_freq)
        if cum_freq + hists[i] > total_n //2:
            break
        else:
            cum_freq += hists[i]
    
    median_val = bins[i] + ((total_n // 2 - cum_freq) / hists[i]) * (bins[1]-bins[0])
    return round(median_val, 4)
```

```python
for i in range(len(col)):
    print("Median", col[i], ":", hist_median(hists[i], bins[i]))
```

Mode of Grouped Data = L + W[(Fm – F1)/( (Fm-F1) + (Fm – F2) )]

where:
- L: Lower limit of modal class
- W: Width of modal class
- Fm: Frequency of modal class
- F1: Frequency of class immediately before modal class
- F2: Frequency of class immediately after modal class


```python
def hist_mode(hists, bins):
    mode_i = list(hists).index(max(hists))

    L = bins[mode_i]
    W = bins[1] - bins[0]
    Fm = hists[mode_i]
    F1 = hists[mode_i-1]
    F2 = hists[mode_i+1]

    mode_val = L + W*((Fm-F1)/((Fm-F1)+(Fm-F2)))
    return round(mode_val, 4)
```

```python
for i in range(len(col)):
    print("Mode", col[i], ":", hist_mode(hists[i], bins[i]))
```
