```python
import numpy as np
import numpy.matlib as matlib
import numpy.linalg as la
```

# Question 1
Consider a matrix A of size nxn(n>=2) and a sub matrix B of size mxm (m<=n).
1. Find the number of submatrix
2. Max, min, average of each submatrix
3. Determinant of each submatrix
4. Inverse of each submatrix

**Assumption: "n" should be even and "m" should be divisible by "n".**

**The element of one submatrix is not matching with other element of submatrix.**

```python
n = int(input("Enter the value of n: "))
# arr = np.mat(input("Enter the matrix as string: "))
arr = np.arange(1, n*n+1).reshape(n, n)
arr = np.array(arr)

m = int(input("Enter the value of m: "))
print(arr)

subarr = []
for i in range(0,n,m):
    temp = []
    for j in range(0,n,m):
        temp.append(arr[i:i+m, j:j+m])
    subarr.append(temp)

subarr = np.array(subarr)
print(subarr)


num_sub = len(subarr)*len(subarr[0])
print("Number of submatrix:", num_sub)


maxarr = []
for a in subarr:
    temp = []
    for b in a:
        temp.append(np.max(b))
    maxarr.append(temp)

maxarr = np.array(maxarr)
print()
print("The max value: ")
print(maxarr)


minarr = []
for a in subarr:
    temp = []
    for b in a:
        temp.append(np.min(b))
    minarr.append(temp)

minarr = np.array(minarr)
print()
print("The min value: ")
print(minarr)


avgarr = []
for a in subarr:
    temp = []
    for b in a:
        temp.append(np.average(b))
    avgarr.append(temp)

avgarr = np.array(avgarr)
print()
print("The average value: ")
print(avgarr)


detarr = []
for a in subarr:
    temp = []
    for b in a:
        temp.append(la.det(b))
    detarr.append(temp)

detarr = np.array(detarr)
print()
print("The determinant value: ")
print(detarr)


invarr = []
for a in subarr:
    temp = []
    for b in a:
        try:
            temp.append(la.inv(b))
        except la.LinAlgError:
            temp.append(np.zeros((m,m)))
    invarr.append(temp)

invarr = np.array(invarr)
print()
print("Null Matrices are for non-invertible")
print("The inverse matrices: ")
print(invarr)
```
