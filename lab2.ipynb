```python
def fibon(n):
    if n == 0 or n == 1:
        return n;
    else:
        return fibon(n-1)+fibon(n-2)

n = int(input("Fibonacci series upto n values: "))
if n <= 0:
    print("Invalid input")
    exit()

for i in range(n):
    print(fibon(i), end=" ")
print()
```

```python
def pat(a, i):
    if a == 1:
        string = "*"*i+" "*(5-i)
    elif a == 2:
        if i == 0:
            return " "*9
        string = " "*(5-i)
        for j in range(1,i):
            string += str(j)
        for j in range(i, 0, -1):
            string += str(j)
        string += " "*(5-i)
    elif a == 3:
        string = "PYTHON"[:i+1]+" "*(6-i)
    elif a == 4:
        string = str(i)*i
    
    return string
```

```python
for x in range(6):
    for y in range(1,5):
        print(pat(y, x), end=" ")
    print()
```
