```python
def to_dec(bin_str):
    dec = 0
    mul = 1
    if "." in bin_str:
        bin_int = abs(int(bin_str.split(".")[0]))
        bin_pnt = bin_str.split(".")[1]
    else:
        bin_int = abs(int(bin_str))
        bin_pnt = ""
    while bin_int>0:
        dec += bin_int%10 *mul
        bin_int = bin_int//10
        mul *= 2
    mul = 0.5
    for b in bin_pnt:
        dec += int(b)*mul
        mul *= 0.5
    if bin_str[0] == "-":
        dec = -dec
    return dec

bin_str = input("Enter a binary number: ")
if set(bin_str).issubset({'0', '1', '-', '.'}):
    print("Decimal:", to_dec(bin_str))
else:
    print("Invalid Binary")
```

```python
def to_bin(dec_str, prec):
    bin_str = ""
    if dec_str[0] == '-':
        bin_str += "-"
        dec_str = dec_str[1:]
    if "." in dec_str:
        dec_int = int(dec_str.split(".")[0])
        dec_pnt = float("0."+dec_str.split(".")[1])
    else:
        dec_int = int(dec_str)
        dec_pnt = 0.0
    
    bin_int, mul = 0, 1
    while dec_int > 0:
        bin_int += dec_int%2 * mul
        dec_int = dec_int//2
        mul *= 10
    bin_str += str(bin_int)
    count = 0
    while not dec_pnt.is_integer() and count != prec:
        if count == 0:
            bin_str += "."
        bin_str += str(int(dec_pnt*2))
        dec_pnt = float("0."+str(dec_pnt*2).split(".")[1])
        count += 1
    return bin_str


dec_str = input("Enter a decimal number: ")
prec = int(input("Enter the precision: "))
print("Binary: ", to_bin(dec_str, prec))
```
