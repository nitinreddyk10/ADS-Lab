## Question 1

Consider a one dimensional table having 20 integer elements ranging from 100 to 200. 

1. Construct the table.
2. Save the table and load the table.
3. Display the shape,size and dimension of table
4. Find min, max, mean ,median, and  standard deviation. Show the result in Tabular form 


```python
ser = pd.Series(np.random.randint(100,201, size=(20)))

fname = 'q1_file.csv'
ser.to_csv(path+fname)

new_ser = pd.read_csv(path+fname)
new_ser = pd.Series(new_ser.iloc[:, -1])
print("Table: ")
print(new_ser)

print()
print("Shape: ", new_ser.shape)
print("Size: ", new_ser.size)
print("Dimension: ", new_ser.ndim)
print()

stats = {
    "Minimum": np.min(new_ser),
    "Maximum": np.max(new_ser),
    "Mean": np.average(new_ser),
    "Median": np.median(new_ser),
    "Std. Deviation": np.std(new_ser)
}


stats_ser = pd.Series(stats)
print(stats_ser)
```

## Question 2

Create two tables Studentinfo and Facultyinfo defined as follows:

    Studentinfo(Studentid, Branch, Semester, Cgpa, Department, DOJ)
    Facultyinfo(Faculty id, Department, Faculty name, DOJ, Salary )

Note: Initially table contain four entries. (take Real data value)


```python
# Data used in the program

# Initial Entries
sDict1 = {
    "Studentid": ["s1", "f3", "c3", "t2"],
    "Branch": ["CS", "CSAI", "EC", "ME"],
    "Semester": [2, 3, 5, 1],
    "CGPA": [8.7, 9.2, 7.9, 10],
    "Department": ["CSE", "CSE", "ECE", "MECH"],
    "DOJ": [dt.datetime(2021, 1, 1), dt.datetime(2021, 9, 3), 
            dt.datetime(2019, 7, 5), dt.datetime(2022, 8, 9)],
}
fDict1 = {
    "Facultyid": ["r4", "w1", "b9", "y5"],
    "Department": ["ECE", "MATH", "CSE", "MECH"],
    "Name": ['Selvaraj','Mishra','Jagadeesh','Arun'],
    "DOJ": [dt.datetime(2008, 3, 20), dt.datetime(2010, 10, 31), 
            dt.datetime(2012, 1, 7), dt.datetime(2013, 2, 1)],
    "Salary": [200000,190000,300000,175000]
}

# After 6 Months
sDict2 = {
    "Studentid": ["f1", "s3", "g2"],
    "Branch": ["EC", "ME", "SM"],
    "Semester": [1, 1, 1],
    "CGPA": [5.4, 7.4, 6.2],
    "Department": ["ECE", "MECH", "MECH"],
    "DOJ": ["2023-02-04"]*3,
}
fDict2 = {
    "Facultyid": ["b2", "e5", "i3", "o7"],
    "Department": ["CSE", "ECE", "MECH", "ECE"],
    "Name": ['Siddesh','Julia','Kiran','Loknath'],
    "DOJ": ["2023-02-12"]*4,
    "Salary": [250000,120000,100000,140000]
}

# After 3 Months
sDict3 = [2, 4, 6] # Row index of students left
fDict3 = {
    "Facultyid": ["d1", "g6", "r2", "l4", "p8"],
    "Department": ["MECH", "CSE", "MATH", "CSE", "ECE"],
    "Name": ['Sharma','Rahul','John','Basheer','Sneha'],
    "DOJ": ["2023-05-20"]*5,
    "Salary": [210000,155000,160000,230000,90000]
}

# Address Column for both tables
sAddr = ["Chennai", "Bengaluru", "Delhi", "Hyderabad"]
fAddr = ["Mysuru", "Raipur", "Chennai", "Allahabad", 
         "Jaipur", "Kolkata", "Mumbai", "Nagpur",
         "Chennai", "Kochi", "Chennai", "Indore"]
```

#### Part: a, b, c
- Construct the table
- Save the table and load the table for further processing
- Display the shape, size and dimension of table

```python
Sinfo = pd.DataFrame(sDict1, index=range(1,5))
Finfo = pd.DataFrame(fDict1, range(1,5))

fnames = ["student.csv", "faculty.csv"]
Sinfo.to_csv(path+fnames[0])
Finfo.to_csv(path+fnames[1])


Studentinfo = pd.read_csv(path+fnames[0], index_col=[0])
Facultyinfo = pd.read_csv(path+fnames[1], index_col=[0])

print(Studentinfo)
print()
print("Shape: ", Studentinfo.shape)
print("Size: ", Studentinfo.size)
print("Dimension: ", Studentinfo.ndim)

print('\n')

print(Facultyinfo)
print()
print("Shape: ", Facultyinfo.shape)
print("Size: ", Facultyinfo.size)
print("Dimension: ", Facultyinfo.ndim)

print('\n')
```

#### Part: d

```python
print("Tables after 6 months")

Studentinfo = Studentinfo.append(pd.DataFrame(sDict2, index=range(5,8)))
Facultyinfo = Facultyinfo.append(pd.DataFrame(fDict2, index=range(5,9)))

print()
print(Studentinfo)
print('\n')
print(Facultyinfo)

print('\n')
```

#### Part: e

```python
print("Tables after 3 months")

Studentinfo = Studentinfo.drop(sDict3)
Studentinfo.index = range(1,5)
Facultyinfo = Facultyinfo.append(pd.DataFrame(fDict2, index=range(9,13)))

print()
print(Studentinfo)
print('\n')
print(Facultyinfo)

print('\n')
```
